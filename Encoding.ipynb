```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
```

# ONE ENCODER(Nominal)

```python
data=pd.read_csv("Students.csv")
data
```

```python
sp=pd.DataFrame(data['Sports Involvement'])
```

```python
data['Age'].unique()
```

```python
df=OneHotEncoder(drop='first').fit_transform(sp).toarray()
```

```python
SP=pd.DataFrame(df,columns=['Sports Involvement Yes'])
```

```python
data.info()
```

```python
data["Sports Involvement Yes"]=SP
```

```python
data.info()
```

```python
data.drop(["Sports Involvement"],axis=1,inplace=True)
```
