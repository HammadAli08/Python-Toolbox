# Ordinal Encoding with MAPPING

```python
Data1=pd.read_csv("factory_worker_data.csv")
```

```python
Data1.head(2)
```

```python
Data1['Department'].unique()
```

```python
ord={'QC':0, 'Assembly':1, 'Packaging':2, 'Logistics':3, 'Maintenance':4}
```

```python
Dep=Data1['Department'].map(ord)
```

```python
Data1['Department']=Dep
```

```python
Data1
```

```python
from sklearn.impute import SimpleImputer
```

```python
Si=SimpleImputer(strategy='mean')
```
