# Ordinal Encoding(In which the variables are connected with each other)

```python
# Ordinal Coding of Depatment of Employes data
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
Data['Department'].unique()
qr=[['QC', 'Assembly', 'Packaging', 'Logistics', 'Maintenance']]
qr
```

```python
oe=OrdinalEncoder(categories=qr)
DF=pd.DataFrame(Data['Department'])
```

```python
oe.fit(DF)
```

```python
oe.transform(DF)
```

```python
Data['Department'] = oe.transform(DF)
```

```python
Data
```

```python
oe1=OrdinalEncoder()
W_df=pd.DataFrame(Data['WorkerID'])
```

```python
oe1.fit(W_df)
```

```python
oe1.transform(W_df)
```

```python
Data['WorkerID']=oe1.transform(W_df)
```
