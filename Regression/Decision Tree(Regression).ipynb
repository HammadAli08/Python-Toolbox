```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("/home/hammadali08/Vidoes/CSV file/Salary Data.csv")
data.drop(columns=['Gender','Education Level','Job Title'],inplace=True)
data
```

```python
data.fillna(data.mean(),inplace=True)
```

```python
data.isnull().sum()
```

```python
sns.pairplot(data)
plt.show()
```

```python
x=data.iloc[:,:-1]
y=data['Salary']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.tree import DecisionTreeRegressor,plot_tree
```

```python
dtc=DecisionTreeRegressor()
```

```python
dtc.fit(x_train,y_train)
```

```python
dtc.score(x_test,y_test),dtc.score(x_train,y_train)
```

It is Overfitted

```python
plt.figure(figsize=(15,10))
plot_tree(dtc)
plt.show()
```

```python
for i in range(1,10):
    dtc1=DecisionTreeRegressor(max_depth=i)
    dtc1.fit(x_train,y_train)
    print(i,dtc1.score(x_test,y_test),dtc1.score(x_train,y_train))
```

```python
dtc1=DecisionTreeRegressor(max_depth=4)
```

```python
dtc1.fit(x_train,y_train)
```

```python
dtc1.score(x_test,y_test),dtc1.score(x_train,y_train)
```
