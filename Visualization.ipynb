```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
a=[2,3,4,5]
b=[6,7,8,9]
c=[10,11,5,10]
num1=list(np.array([12,14,19,23,29,45,39]))
num2=list(np.array([45,65,23,43,12,23,45]))
no=list(np.array([1,2,9,7,5,6,10]))
```

```python
plt.bar(a,b,color='b',width=0.4,linestyle='-',linewidth=2,edgecolor='k',alpha=0.7,label='Bar chart Label')
plt.bar(a,c,color='g',width=0.4,linestyle='-',linewidth=2,edgecolor='k',alpha=0.7,label='Bar chart Label 2')
plt.title('Bar Chart')
plt.xlabel('Value')
plt.ylabel('Count')
plt.legend()
plt.show()
```

```python
plt.plot(b,c,color='blue')
plt.scatter(b,c)
plt.show()
```

```python
mean=np.mean(num1)
mean2=np.mean(num2)
median=np.median(num1)
median2=np.median(num2)
```

```python
plt.figure(figsize=(10,3))
plt.scatter(num1,no,color="Green",label="num1")
(plt.scatter(num2,no,color="red",label="num2"))
plt.plot([mean]*7,no,color="blue",label="mean")
plt.plot([mean2]*7,no,color="yellow",label="mean")
plt.plot([median]*7,no,color="yellow",label="median")
plt.plot([median2]*7,no,color="brown",label="median2""")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,3))
plt.title("Movie Rating")
plt.scatter(num2,no,color="Green",label="num1")
(plt.scatter(num1,no,color="red",label="num2"))
plt.plot(num2,no)
plt.plot(num1,no)
plt.xlabel("Numbers")
plt.ylabel("Sample")
plt.show()
```

```python
size=list(np.array([30,40,15,15]))
label=list(np.array(['A','B','C','D']))
plt.pie(size,labels=label,autopct='%1.1f%%')
plt.show
```

```python
data=np.random.randn(10000)
plt.hist(data,bins=100,edgecolor="red")

plt.title("Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
data=list(np.random.normal(5,10,100))
data2=list(np.random.normal(25,10,100))
plt.figure(figsize=(6,4))
sns.boxplot([data,data2], patch_artist=True)
plt.title("Box Plot")
plt.xlabel("values")
plt.ylabel("values")
plt.xticks([0, 1], ["Data1", "Data2"])
plt.show()
```

```python
data=list(np.random.normal(5,10,100))
data2=list(np.random.normal(25,15,100))
plt.figure(figsize=(6,4))
sns.violinplot([data,data2])
plt.title("violin plot")
plt.xlabel("values")
plt.ylabel("Points")
plt.xticks([0, 1], ["Data1", "Data2"])
plt.show()
```

```python
plt.show()
plt.fill_between(num1, num2, no, color="blue", alpha=0.3)
plt.title("Fill Between")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```
