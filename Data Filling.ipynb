```python
DATA=pd.read_csv('CSV Files/car_data.csv')
```

```python
DATA['Odometer (KM)'].dropna().sample()
```

```python
DATA['Odometer (KM)'].dropna().sample(DATA['Odometer (KM)'].isnull().sum())
```

```python
random=DATA['Odometer (KM)'].dropna().sample(DATA['Odometer (KM)'].isnull().sum(),random_state=0)
```
