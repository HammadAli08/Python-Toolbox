```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('car_data.csv')
data.head(3)
```

```python
data.fillna({'Odometer (KM)': data['Odometer (KM)'].mean()},inplace=True)
```

```python
data.fillna({'Price': data['Price'].mean()},inplace=True)
```

```python
data.isnull().sum()
```

```python
from sklearn.preprocessing import MinMaxScaler
mms=MinMaxScaler()
```

```python
mms.fit(data[['Odometer (KM)']])
```

```python
MinMaxScaler=mms.transform(data[['Odometer (KM)']])
```

```python
data['Odometer (KM) MMS']=MinMaxScaler
```

# Min Max scalar reduces data in between 0 and 1. But the nature of data remains same

```python
data.describe()
```

```python
plt.subplot(2,1,1)
plt.title('Odometer (KM) Before')
sns.distplot(data['Odometer (KM)'])
plt.show()

plt.subplot(2,1,2)
plt.title('Odometer (KM) After')
sns.distplot(data['Odometer (KM) MMS'])
plt.show()
```

```python
mms.fit(data[['Price']])
```

```python
mms.transform(data[['Price']])
```

```python
data['Price MMS']=mms.transform(data[['Price']])
```

```python
data.describe()
```

```python
plt.subplot(2,1,1)
plt.title('Price Before')
sns.distplot(data['Price'])
plt.show()
plt.subplot(2,1,2)
plt.title('Price After')
sns.distplot(data['Price'])
```

The NATURE of data remains the same
