```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("/home/hammadali08/Vidoes/CSV file/Social_Network_Ads.csv")
data.drop(columns=['User ID','Gender'],inplace=True)
data
```

```python
sns.scatterplot(x='Age',y='EstimatedSalary',data=data,hue='Purchased')
```

```python
x=data.iloc[:,:-1]
y=data['Purchased']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=70)
```

```python
from sklearn.svm import SVC
```

```python
svc=SVC(kernel='poly')
svc.fit(x_train,y_train)
```

```python
 svc.score(x_test,y_test),svc.score(x_train,y_train)
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=svc,legend=2)
```

# For Polynomial Data

```python
data1=pd.read_csv('/home/hammadali08/Vidoes/CSV file/polynomial_logistic_dataset.csv')
data1
```

```python
x1=data1.iloc[:,:-1]
y1=data1['Output']
```

```python
x1_train,x1_test,y1_train,y1_test=train_test_split(x1,y1,test_size=0.25,random_state=13)
```

```python
sv=SVC(kernel='rbf')
sv.fit(x1_train,y1_train)
```

```python
sv.score(x1_test,y1_test),sv.score(x1_train,y1_train)
```

```python
plot_decision_regions(x1.to_numpy(),y1.to_numpy(),clf=sv,legend=2)
```
