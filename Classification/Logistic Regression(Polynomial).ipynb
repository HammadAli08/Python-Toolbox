```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("/home/hammadali08/Vidoes/CSV file/polynomial_logistic_dataset.csv")
data
```

```python
sns.scatterplot(x="x1",y="x2",data=data,hue='Output')
plt.legend(loc='upper right')
plt.show()
```

```python
x=data.iloc[:,:-1]
y=data['Output']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lg=LogisticRegression()
```

```python
lg.fit(x_train,y_train)
```

```python
lg.score(x_test,y_test)
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=lg,colors="blue,red",legend=2)
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("Decision Boundary")
```

# Now trying this dataset with Polynomial Feaures

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
pf=PolynomialFeatures(degree=3)
```

```python
X=data.iloc[:,:-1]
Y=data['Output']
```

```python
pf.fit(X)
```

```python
X=pd.DataFrame(pf.transform(X))
X
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=48 )
```

```python
lg.fit(X_train,Y_train)
```

```python
lg.score(X_test,Y_test)*100
```
