```python
data=list(np.random.normal(5,10,100))
data2=list(np.random.normal(25,10,100))
plt.figure(figsize=(6,4))
sns.boxplot([data,data2], patch_artist=True)
plt.title("Box Plot")
plt.xlabel("values")
plt.ylabel("values")
plt.xticks([0, 1], ["Data1", "Data2"])
plt.show()
```

```python
data1 = np.random.normal(60, 5, 100)
data2 = np.random.normal(45, 8, 100)

plt.figure(figsize=(8, 5))
plt.boxplot([data1, data2], tick_labels=['Group 1', 'Group 2'], patch_artist=True)
plt.title("Boxplot of Two Groups")
plt.ylabel("Values")
plt.show()
```

```python
#GPT ki aulaaad
data = list(np.random.normal(5, 10, 100))
data2 = list(np.random.normal(25, 10, 100))
# ✅ Fix: pass data as list of lists, and use correct 'names' for x-tick labels
sns.boxplot(data=[data, data2])
plt.title("Box Plot")
plt.xlabel("Groups")
plt.ylabel("Values")
plt.xticks([0, 1], ["Data1", "Data2"])  # ✅ Correctly label the two boxplots
plt.show()
```

```python
data=list(np.random.normal(5,10,100))
data2=list(np.random.normal(25,10,100))
plt.figure(figsize=(6,4))
sns.violinplot([data,data2])
plt.title("violin plot")
plt.xlabel("values")
plt.ylabel("Points")
plt.xticks([0, 1], ["Data1", "Data2"])
plt.show()
```
