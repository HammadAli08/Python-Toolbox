# Scalling is used to Shrink the data
# Types: Standardization and Normalization

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
data=pd.read_csv('CSV Files/car_data.csv')
data
```

```python
data.fillna({'Price': data['Price'].mean()}, inplace = True)
```

```python
data.fillna({'Odometer (KM)': data['Odometer (KM)'].mean()}, inplace = True)
```

```python
data.isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
```

```python
ss.fit(data[['Odometer (KM)']])
```

```python
ss.transform(data[['Odometer (KM)']])
```

```python
data['Odometer (KM) SS']=ss.transform(data[['Odometer (KM)']])
```

```python
data
```

```python
plt.subplot(2,1,1)
plt.title('Odometer (KM) Before')
sns.distplot(data['Odometer (KM)'])
plt.show()

plt.subplot(2,1,2)
plt.title('Odometer (KM) After')
sns.distplot(data['Odometer (KM) SS'])
plt.show()
```

```python
ss.fit(data[['Price']])
```

```python
ss.transform(data[['Price']])
```

```python
data["Price SS"]=ss.transform(data[['Price']])
```

```python
plt.subplot(2,1,1)
plt.title('Price Before')
sns.distplot(data['Price'])
plt.show()

plt.subplot(2,1,2)
plt.title('Price After')
sns.distplot(data['Price SS'])
plt.show()
```
